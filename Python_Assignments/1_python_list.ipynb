1. Find the Sum: Calculate the total of all numerical elements in a list.

```python
l= [1,23,4,5,6,788,876,1]
# print(sum(l))
```

```python
l= [1,23,4,5,6,788,876,1]
sum = 0
for i in l:
    sum+= i
print(sum)
```

```python
l= [1,23,4,5,6,788,876,1]
import functools
a= functools.reduce(lambda x,y: x+y,l)
print(a)
```

2. Find the Largest/Smallest: Identify the maximum and minimum values without using built-in max() or min().

```python
l= [1,23,4,5,6,788,876,1]
l.sort()
print(f"maximum number is- {l[-1]} and minimum number is {l[0]}")
```

3. Check for Existence: Determine if a specific element exists within a list.

```python
l= [1,23,4,5,6,788,876,1]
a = int(input('enter element to check'))

if a in l:
    print(a,'is in list')
else:
    print(a,'is not in list')
```

4. Reverse a List: Flip the order of elements (bonus: do it in-place and using slicing [::-1]).

```python
l = ['my','name','is','lakhan']
print(l[::-1])
```

```python
for i in range(len(l)-1,-1,-1):
    print(l[i], end=',')
```

5. Count Occurrences: Count how many times a specific element appears in the list.

```python
l= [1,23,4,5,6,788,876,1]
a= int(input('enter number to check frequecy'))
counter = 0
for i in l:
    if i ==a:
        counter+=1
print(f"{a} appeared {counter} times")
```

8. List Slicing: Extract the first half of a list and the last half into two separate variables.

```python
l= [1, 4, 5, 6, 876, 788, 23,981]
a = l[:(int(len(l)/2))]
b = l[(int(len(l)/2)):]
print(a)
print(b)
len(l)

```

10. Filter Evens/Odds: Use a list comprehension to create a list of only even (or odd) numbers.

```python
l= [1, 4, 5, 6, 876, 788, 23,981]
even = [i for i in l if i%2 == 0]
odd = [i for i in l if i%2 == 1]
print(even)
print(odd)
```

11. Merge and Sort: Combine two sorted lists into one single sorted list.

```python
a= [1, 4, 5, 6]
b= [876, 788, 23, 981]
c= sorted(a+b)
print(c)
```

12. Cumulative Sum: Transform a list $[a, b, c]$ into $[a, a+b, a+b+c]$.

```python
list= [1, 4, 5, 6, 23, 788, 876, 981]
sum = 0
l=[]
for i in list:
    sum= sum+i
    l.append(sum)
print(l)
```
