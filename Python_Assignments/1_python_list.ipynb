1. Find the Sum: Calculate the total of all numerical elements in a list.

```python
l= [1,23,4,5,6,788,876,1]
# print(sum(l))
```

```python
l= [1,23,4,5,6,788,876,1]
sum = 0
for i in l:
    sum+= i
print(sum)
```

```python
l= [1,23,4,5,6,788,876,1]
import functools
a= functools.reduce(lambda x,y: x+y,l)
print(a)
```

2. Find the Largest/Smallest: Identify the maximum and minimum values without using built-in max() or min().

```python
l= [1,23,4,5,6,788,876,1]
l.sort()
print(f"maximum number is- {l[-1]} and minimum number is {l[0]}")
```

```python
l= [1,23,4,5,6,788,876,1]
import functools
a= functools.reduce(lambda x,y: x>y,l)
print(a)
```
