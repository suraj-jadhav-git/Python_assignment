7. Write a python program to find the longest words.


```python
with open('/content/sample.txt', 'r') as f:
  l= list(f.readlines())

  a= l[0] # here we extract string from list
  b= (a.split())
  # print(b)
  l1=[]
  for i in b:
    if len(i) > 5: #assumption word with more than 5 letters is cosidered longest  word
      l1.append(i)
  print(l1)
```
