 Challenge: Student Marks Analyzer

Create a Python program that allows a user to input student names along with their marks and then calculates useful statistics.
Your program should:
1. Let the user input multiple students with their marks (name + integer score).
2. After input is complete, display:
   - Average marks
   - Highest marks and student(s) who scored it
   - Lowest marks and student(s) who scored it
   - Total number of students

Bonus:
- Allow the user to enter all data first, then view the report
- Format output clearly in a report-style layout
- Prevent duplicate student names


```python
class Student:
  def __init__(self):
    self.d={}
    self.pannel()

  def pannel(self):
    a = input('''
    1. Add Student
    2. Average Marks
    3. Highest mark student
    4. Lowest mark student
    5. Total Number of student
    6. Exit()
    ''')
    if a == '1':
      self.addStudent()
    elif a == '2':
      self.avgMark()
    elif a == '3':
      self.maxMark()
    elif a == '4':
      self.minMark()
    elif a == '5':
      self.studentCount()
    elif a == '6':
      print('ThankYou!')
      exit()

  def addStudent(self):
    name = input('Enter Student Name: ')
    mark = int(input('Enter Student Marks: '))
    self.d[name]=mark
    self.pannel()

  def avgMark(self):
    l = list(self.d.values())
    avg =sum(l)/len(l)
    print('Average Marks : ',avg)
    self.pannel()

  def maxMark(self):
    b = max(self.d.values())
    for i in self.d:
      if self.d[i] == b:
        print('name: ',i,'marks: ',self.d[i])
    self.pannel()

  def minMark(self):
    c = min(self.d.values())
    for i in self.d:
      if self.d[i] == c:
        print('name: ',i,'marks: ',self.d[i])
    self.pannel()

  def studentCount(self):
    l = list(self.d.values())
    print('Total Number of Student: ',len(l))
    self.pannel()

o = Student()

```
