6. Write a Python program to create a class representing a stack data structure. Include methods for pushing and popping elements.


```python

class Stack:
  def __init__(self):
    self.l=[1,2,3,4,5,6]
    self.function()

  def function(self):
    function=input('enter what would you like to do push/pull?')
    if function =='push':
      self.push()
    else:
      self.pull()

  def push(self):
    add=int(input('enter number to add'))
    self.l.append(add)
    print('added',add,'to',self.l)
    self.function()

  def pull(self):
    if self.l ==[]:
      print('stack is empty now')
    else:
      re=self.l[-1]
      self.l.remove(re)
      print('removed last item',re,'from',self.l)
      self.function()

obj=Stack()
```

10. Write a Python program to create a class representing a queue data structure. Include methods for enqueueing and dequeueing elements.

```python

class Queue:
  def __init__(self):
    self.l=[1,2,3,4,5,6]
    self.function()

  def function(self):
    function=input('enter what would you like to do push/pull?')
    if function =='push':
      self.enqueueing()
    else:
      self.dequeueing()

  def enqueueing(self):
    add=int(input('enter number to add'))
    self.l.append(add)
    print('added',add,'to',self.l)
    self.function()

  def dequeueing(self):
    if self.l ==[]:
      print('stack is empty now')
    else:
      re=self.l[0]
      self.l.remove(re)
      print('removed last item',re,'from',self.l)
      self.function()

obj=Queue()
```
