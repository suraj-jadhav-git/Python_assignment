6. Write a Python program to create a class representing a stack data structure. Include methods for pushing and popping elements.


```python

class Stack:
  def __init__(self):
    self.l=[1,2,3,4,5,6]
    self.function()

  def function(self):
    function=input('enter what would you like to do push / pop / exit?')
    if function =='push':
      self.push()
    elif function =='pop':
      self.pop()
    elif function == 'exit':
      print('thanks')
      exit()
    else:
      self.function()

  def push(self):
    add=int(input('enter number to add'))
    self.l.append(add)
    print('added',add,'to',self.l)
    self.function()

  def pop(self):
    if self.l ==[]:
      print('stack is empty now')
    else:
      re=self.l[-1]
      self.l.remove(re)
      print('removed last item',re,'from',self.l)
      self.function()

obj=Stack()
```
